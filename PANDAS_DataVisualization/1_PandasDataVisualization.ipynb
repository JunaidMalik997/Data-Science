# Pandas built in Data VIsualization

```python
#we will go through pandas own built in capability that is built off of matplotlib, but it allows us to 
#call data visualizations drectly of a dataframe.
```

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
dfa=pd.read_csv('df1')
```

```python
dfa.head()
```

```python
df1=pd.read_csv('df1',index_col=0)  #time series index
```

```python
df1.head()
```

```python
df2=pd.read_csv('df2')    #non time series index
```

```python
df2.head()
```

```python
#imagine we want to get a histogram of all the values in the A column of df1
df1['A'].hist()
```

```python
df1['A'].hist(bins=30)
```

```python
#the style above is not good so to fix it we can import seaborn
import seaborn as sns
```

```python
df1['A'].hist(bins=30)
#we can use seaborn styles and pandas calls will look like them
```

```python
#pandas has several plot types built into it. Pretty much all the plots can be done through the command off of the dataframe
df1['A'].plot(kind='hist')
#this is one way of calling the plot
```

```python
#another way of calling plots using pandas is as follows--> call the plot directly
df1['A'].plot.hist()
```

```python
df2.head()
```

```python
df2.plot.area()
#this area plot just shows the area of the values
```

```python
df2.plot.area(alpha=0.4)
#alpha argument for more transparency
```

```python
df2.plot.bar()
#it just takes in the index value basically as a category(should be categorical index) it will list the value per each column
```

```python
df2.plot.bar(stacked=True)
```

```python
df1.head()
```

```python
df1.plot.line(y='B')
```

```python
#since it is built of matplotlib we can add matplotlib arguments to it such as figsize
df1.plot.line(y='B',figsize=(12,3),lw=1)
```

```python
df1.plot.scatter(x='A',y='B')
```

```python
df1.plot.scatter(x='A',y='B',c='C')
#c(color) will be equal of column C
#so now we are showing 3 levels of information
```

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
#instead of c color people mostly use s(size)
df1.plot.scatter(x='A',y='B',s=df1['C']*100)
#plot of A vs B and the size indicates how much there size value was as compared to each other.
```

```python
df2.plot.box()
```

```python
df=pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
```

```python
df.head()
```

```python
#since it is by variat data(2 variables) a and b we can say that
df.plot.hexbin(x='a',y='b')
#essentially a scatter plot instead we have hexagonal bins
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='coolwarm')
```

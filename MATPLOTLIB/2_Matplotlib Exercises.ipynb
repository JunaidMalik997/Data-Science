```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y,'red')
ax2.plot(x,y,'red')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.4,0.4])

ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax1.set_xlim([0,100])
ax1.set_ylim([0,10000])

ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,'blue',lw=3,ls='--')
axes[1].plot(x,z,'red',lw=3,ls='-')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2, figsize=(10,4))

axes[0].plot(x,y,'blue',lw=3,ls='--')
axes[1].plot(x,z,'red',lw=3,ls='-')
```
